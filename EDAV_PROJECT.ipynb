```python
#4Q
import pandas as pd
df = pd.read_excel('event_tickets.xlsx')
df = df[df['ticket_price'] >= 1]

correlation = df['ticket_price'].corr(df['tickets_sold'])
print("Correlation between ticket price and tickets sold:", correlation)
```

```python
#5Q
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('event_tickets.xlsx')
df = df[df['ticket_price'] >= 1]
df['payment_mode'] = df['payment_mode'].fillna('Unknown')

# ---- Bar chart: average tickets sold by day_of_week ----
avg_tickets_day = df.groupby('day_of_week')['tickets_sold'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,5))
avg_tickets_day.plot(kind='bar', color='skyblue')
plt.title('Average Tickets Sold by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Tickets Sold')
plt.xticks(rotation=45)
plt.show()

# ---- Scatter plot: ticket_price vs tickets_sold ----
plt.figure(figsize=(8,5))
plt.scatter(df['ticket_price'], df['tickets_sold'], color='orange', alpha=0.6)
plt.title('Ticket Price vs Tickets Sold')
plt.xlabel('Ticket Price')
plt.ylabel('Tickets Sold')
plt.show()
```
